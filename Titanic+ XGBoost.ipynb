```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/home/mikaeil/Desktop/Dataset/titanic/train.csv')
```

```python
df_new=pd.read_csv('/home/mikaeil/Desktop/Dataset/titanic/test.csv')
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer
from xgboost import XGBClassifier
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
cols = ['Parch' , 'Fare', 'Embarked', 'Sex', 'Name', 'Age']
```

```python
X = df[cols]
y = df['Survived']
```

```python
X_new = df_new[cols]
```

```python
imp_constant = SimpleImputer(strategy='constant', fill_value='missing')
ohe = OneHotEncoder()
```

```python
imp_ohe = make_pipeline(imp_constant, ohe)
vect = CountVectorizer()
imp = SimpleImputer()
```

```python
ct = make_column_transformer(
(imp_ohe, ['Embarked', 'Sex']),
(vect, 'Name'),
(imp, ['Age', 'Fare']),
remainder='passthrough')
```

```python
xgb_cla = XGBClassifier(ocolsample_bytree = 0.3, n_jobs=4)
```

```python
pipe = make_pipeline(ct, xgb_cla)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X, y, cv=5, scoring='accuracy').mean()
```

```python
params = {}
params['xgbclassifier__n_estimators']=[10, 50, 100]
params['xgbclassifier__learning_rate']=[0.1, 0.5, 1.0,]
params['xgbclassifier__max_depth']=[3, 10, 15]
params['xgbclassifier__alpha']=[2, 5, 10,]
params['columntransformer__pipeline__onehotencoder__drop'] = [None, 'first']
params['columntransformer__countvectorizer__ngram_range'] = [(1, 1), (1, 2)]
params['columntransformer__simpleimputer__add_indicator'] = [False, True]
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
```

```python
grid.fit(X,y);
```

```python
grid.best_score_
```

```python
grid.best_params_
```
